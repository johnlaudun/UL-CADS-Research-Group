```python
# Import necessary modules from NLTK
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize

# Define the file path
file_path = '/Users/ammritolalroy/Downloads/OpenAI.txt'

# Read the content of the file
with open(file_path, 'r') as file:
    content = file.read()

# Tokenize the text content
tokens = word_tokenize(content)

# Calculate frequency distribution of tokens
fdist1 = FreqDist(tokens)

# Print frequency distribution
print(fdist1)

# Optionally, you can view the most common words
print(fdist1.most_common(10))  # Prints top 10 most common words in your file
```
